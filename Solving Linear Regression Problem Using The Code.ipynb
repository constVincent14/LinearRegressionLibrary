<h3>Example 01</h3>
<p>In this example, the feature it "tekanan" and the target is "tegangan," we want to fit a linear regression to find the correct parameters from the function <b>h(x) = theta_zero + theta_one*x</b>, so that we have a model capable of predicting "tegangan" from x-value "tekanan"</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import regression as rs

tekanan = [0.0, 17.24, 34.47, 51.71, 68.95, 86.18, 103.42]
tegangan = [0.81, 0.98, 1.07, 1.16, 1.26, 1.34, 1.45]

model = rs.build_model(tekanan, tegangan)
model.fit_model()
params = model.show_params()
cost = model.calc_cost()

theta_zero = params[0]
theta_one = params[1]
```

```python
tegangan_pred = []
for i in tekanan:
    tegangan_pred.append(theta_zero + (theta_one*i))

%matplotlib inline
plt.scatter(tekanan, tegangan, color='red', label='DATA')
plt.plot(tekanan, tegangan_pred, color='black', label='LEAST SQUARE')
plt.xlabel('TEKANAN')
plt.ylabel('TEGANGAN')
plt.grid(color='#333', ls='--', lw=0.5)
plt.legend()
plt.show()
```

<h3>Example 02</h3>
<p>Here's another example of using the code, in this example the feature is "putaran" and the target we want our model to predict is "pulsa."</p>

```python
with open('from_XOR_clockwise.txt', 'r') as file:
    container = [each.rstrip() for each in file]
    file.close()

temp = []
for index, value in enumerate(container):
    if index==0:
        continue
    temp.append(value.split("\t"))

putaran = []
pulsa = []
for each in temp:
    putaran.append(float(each[0]))
    pulsa.append(float(each[-1])) 

print("PUTARAN: {}".format(putaran), 
      "PULSA: {}\n".format(pulsa), sep='\n\n')



regressor = rs.build_model(putaran, pulsa)
regressor.fit_model()
params = regressor.show_params()
cost = regressor.calc_cost()
```

```python
#result

y_pred = [(params[0]+(params[1]*value)) for value in putaran]
plt.scatter(putaran, pulsa, color='red', label="DATA")
plt.plot(putaran, y_pred, color='black', label='Least Square')
plt.xlabel('Putaran')
plt.ylabel('Pulsa')
plt.grid(color='#333', ls='--')
```
